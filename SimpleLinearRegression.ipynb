# **Simple Linear Regression**
## Predicts a linear relationship between a feature and a label
## Y = c + mX (LINEAR)
## [Further Information at Wikipedia](https://en.wikipedia.org/wiki/Simple_linear_regression)

## Obtain Linear Regression 

```python
import SimpleLinearRegression as slr
linreg  = slr.LinReg()
```

## Obtain independent and dependent data

```python
names = linreg.getData()
print("Feature: ", names[0], " Label: ", names[1])
```

## Split Data to form training set and testing set

```python
splitData = linreg.splitData()
print("Feature Train: ", len(splitData[0])," Label Train: ", len(splitData[2]), " Feature Test: ", len(splitData[1]),  "Label Test: ", len(splitData[3]))
```

## Form the Model and make predictions

```python
linreg.formModel()
scores = linreg.makePredictions()
print("Training r2 Score: ", scores[0], " Testing r2 Score: ", scores[1])
```

## Plot the Results

```python
linreg.plotTrainingResults()
linreg.plotTestResults()
```
